```
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(alpha=1, x=1, bias=1):
  return alpha / (bias + np.e**-x)

sigmoid()
```

```
X = range(-100, 100)
points = [sigmoid(x=i) for i in X]

# modulate the alpha
alpha_labels = []
alpha_funcs = []
for alpha in range(1,10):
  points = [sigmoid(alpha=alpha, x=i) for i in X]
  alpha_labels.append(alpha)
  alpha_funcs.append(points)

print(alpha_funcs, "\n", alpha_labels)
```

```
for r in alpha_labels:
  i = alpha_funcs[r-1]
  r = "alpha = " + str(r)
  plt.scatter(X, i)
  plt.text(120, i[-1], r)
plt.show()
```

```
X = range(-100, 100)
points = [sigmoid(x=i) for i in X]

# modulate the bias

bias_labels = []
bias_funcs = []
for bias in range(1,10):
  points = [sigmoid(x=i, bias=bias) for i in X]
  bias_labels.append(bias)
  bias_funcs.append(points)

for r in bias_labels:
  i = bias_funcs[r-1]
  r = "bias = " + str(r)
  plt.scatter(X, i)
  plt.text(120, i[-1], r)
plt.show()
```
